Indexing

```python
df = pd.read_csv('./../datasets/iris.csv')

print(df['species'])

print(df[['species', 'sepalwidth']])

print(df[2:7])

# تحديد الصفوف وفق شرط
rows = df[(df['species']=='Iris-setosa') & (df['sepallength']==4.7)]
print(len(rows))
print(df.loc[4])
print(df.loc[[4, 5, 6]])
```

Change index column

```python
df.set_index("id", inplace=True)
df.loc[1]
```

Reset index column

```python
df.reset_index("id", inplace=True)
df.loc[1]
```

Get some columns from some rows

```python
print(df.loc[[1, 3, 5], ['species', 'sepalwidth']])
```

Filtering

```python
print(df.loc[df['sepallength'] > 7.5, ['species', 'sepalwidth', 'sepallength']])
```

All rows and some columns

```python
print(df.loc[:, ['species', 'sepalwidth', 'sepallength']])
```

Range of rows

```python
print(df.loc[2:6])
print(df.loc[2:6, ['sepalwidth', 'id']])
```

Filtering by numbering indexes

```python
print(df.iloc[2])
print(df.iloc[[2, 5]])
print(df.iloc[2:5])
print(df.iloc[2, [0, 2]])
```

Iteration over rows

```python
for index, row in df.iterrows():
    print(index, '\n', row, row['species'], '\n', row[['species', 'id']])
```

Iteration over rows as tuples

```python
for row in df.itertuples():
    print(row, '\n', row.species)
```

Iteration over columns names

```python
for column_name in df.columns:
    print(df[column_name], '\n', df[column_name][2])
```

Iteration over columns and their names

```python
for column_name, column in df.iteritems():
    print(column, '\n', column_name, '\n', column.values)
```

Iteration over columns and their names for some columns and rows

```python
for column_name, column in df.loc[[2,5], ['species', 'sepalwidth']].iteritems():
    print(column, '\n', column_name, '\n', column.values)
```
